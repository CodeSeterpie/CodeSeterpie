```python

def addTotalGrSF(df):

    dv_ary = []
    for index, row in df.iterrows():

        dv_ary.append(row['1stFlrSF'] + row['2ndFlrSF'])

    # Seriesに変換
    dv = pd.Series(dv_ary).astype(float)
    # DataFrameと結合
    df['TotalGrSF'] = dv


addTotalGrSF(train)
addTotalGrSF(test)

# 2020/03/28　追加


def addSoldYM(df):

    dv_ary = []
    for index, row in df.iterrows():

        dv_ary.append(str(row['YrSold']) + '{:0>2}'.format(row['MoSold']))

    # Seriesに変換
    dv = pd.Series(dv_ary).astype(int)
    # DataFrameと結合
    df['SoldYM'] = dv


addSoldYM(train)
addSoldYM(test)


def calcMean_LotArea(df, target, dv_count):

    dv_ary = []
    for index, row in df.iterrows():

        dv_ary.append(dv_count[row[target]])

    # Seriesに変換
    dv = pd.Series(dv_ary).astype(float)
    # DataFrameと結合
    df[target + '_LotArea_Mean'] = dv


def addMean_LotArea(df1, df2, target):

    dv_count = df1.append(df2, ignore_index=True)

    # target毎のNLotAreaの平均を算出
    dv_count = dv_count.groupby(target)['LotArea'].mean()

    calcMean_LotArea(df1, target, dv_count)
    calcMean_LotArea(df2, target, dv_count)


addMean_LotArea(train, test, 'Neighborhood')
addMean_LotArea(train, test, 'Condition1')
addMean_LotArea(train, test, 'HouseStyle')
addMean_LotArea(train, test, 'MSSubClass')


def calcStd_LotArea(df, target, dv_count):

    dv_ary = []
    for index, row in df.iterrows():

        dv_ary.append(dv_count[row[target]])

    # Seriesに変換
    dv = pd.Series(dv_ary).astype(float)
    # DataFrameと結合
    df[target + '_LotArea_Std'] = dv


def addStd_LotArea(df1, df2, target):

    dv_count = df1.append(df2, ignore_index=True)

    # target毎のNLotAreaの標準偏差を算出
    dv_count = dv_count.groupby(target)['LotArea'].std()

    calcStd_LotArea(df1, target, dv_count)
    calcStd_LotArea(df2, target, dv_count)


addStd_LotArea(train, test, 'Neighborhood')
addStd_LotArea(train, test, 'Condition1')
addStd_LotArea(train, test, 'HouseStyle')
addStd_LotArea(train, test, 'MSSubClass')


def calcVar_LotArea(df, target, dv_count):

    dv_ary = []
    for index, row in df.iterrows():

        dv_ary.append(dv_count[row[target]])

    # Seriesに変換
    dv = pd.Series(dv_ary).astype(float)
    # DataFrameと結合
    df[target + '_LotArea_Var'] = dv


def addVar_LotArea(df1, df2, target):

    dv_count = df1.append(df2, ignore_index=True)

    # target毎のNLotAreaの分散を算出
    dv_count = dv_count.groupby(target)['LotArea'].var()

    calcVar_LotArea(df1, target, dv_count)
    calcVar_LotArea(df2, target, dv_count)


addVar_LotArea(train, test, 'Neighborhood')
addVar_LotArea(train, test, 'Condition1')
addVar_LotArea(train, test, 'HouseStyle')
addVar_LotArea(train, test, 'MSSubClass')
```

```python

def calcFeatureValue(df, col1, calc, col2):

    dv_ary = []
    for index, row in df.iterrows():
        if calc == 'multi':

            if np.isnan(row[col1]) or np.isnan(row[col2]) \
                   or row[col1] == 0 or row[col2] == 0:
                dv_ary.append(np.log(1))
            else:
                dv_ary.append(np.log(row[col1] * row[col2]))

        elif calc == 'div':
            if np.isnan(row[col2]) or row[col2] == 0:
                dv_ary.append(0)
            else:
                dv_ary.append(row[col1] / row[col2])

        else:
            raise Exception('calcの値が不正です')

    # Seriesに変換
    dv = pd.Series(dv_ary).astype(float)
    # DataFrameと結合
    df[col1 + '_' + calc + '_' + col2] = dv


def addFeatureValue(df1, df2, col1, calc, col2):

    calcFeatureValue(df1, col1, calc, col2)
    calcFeatureValue(df2, col1, calc, col2)


addFeatureValue(train, test, 'LotArea', 'multi', 'LotShape')
addFeatureValue(train, test, 'LotArea', 'div', 'LotShape')
addFeatureValue(train, test, 'LotShape', 'div', 'LotArea')

addFeatureValue(train, test, 'TotalBsmtSF', 'multi', 'BsmtCond')
addFeatureValue(train, test, 'TotalBsmtSF', 'div', 'BsmtCond')
addFeatureValue(train, test, 'BsmtCond', 'div', 'TotalBsmtSF')

addFeatureValue(train, test, 'BsmtQual', 'multi', 'BsmtCond')
addFeatureValue(train, test, 'BsmtQual', 'div', 'BsmtCond')
addFeatureValue(train, test, 'BsmtCond', 'div', 'BsmtQual')

addFeatureValue(train, test, 'GrLivArea', 'multi', 'TotalGrSF')
addFeatureValue(train, test, 'GrLivArea', 'div', 'TotalGrSF')
addFeatureValue(train, test, 'TotalGrSF', 'div', 'GrLivArea')

addFeatureValue(train, test, 'GrLivArea', 'multi', '1stFlrSF')
addFeatureValue(train, test, 'GrLivArea', 'div', '1stFlrSF')
addFeatureValue(train, test, '1stFlrSF', 'div', 'GrLivArea')

addFeatureValue(train, test, 'GrLivArea', 'multi', '2ndFlrSF')
addFeatureValue(train, test, 'GrLivArea', 'div', '2ndFlrSF')
addFeatureValue(train, test, '2ndFlrSF', 'div', 'GrLivArea')

addFeatureValue(train, test, 'Fireplaces', 'multi', 'FireplaceQu')
addFeatureValue(train, test, 'Fireplaces', 'div', 'FireplaceQu')
addFeatureValue(train, test, 'FireplaceQu', 'div', 'Fireplaces')

addFeatureValue(train, test, 'KitchenAbvGr', 'multi', 'KitchenQual')
addFeatureValue(train, test, 'KitchenAbvGr', 'div', 'KitchenQual')
addFeatureValue(train, test, 'KitchenQual', 'div', 'KitchenAbvGr')

addFeatureValue(train, test, 'BedroomAbvGr', 'multi', 'TotRmsAbvGrd')
addFeatureValue(train, test, 'BedroomAbvGr', 'div', 'TotRmsAbvGrd')
addFeatureValue(train, test, 'TotRmsAbvGrd', 'div', 'BedroomAbvGr')

addFeatureValue(train, test, 'PoolQC', 'multi', 'PoolArea')
addFeatureValue(train, test, 'PoolQC', 'div', 'PoolArea')
addFeatureValue(train, test, 'PoolArea', 'div', 'PoolQC')

addFeatureValue(train, test, 'GarageQual', 'multi', 'GarageArea')
addFeatureValue(train, test, 'GarageQual', 'div', 'GarageArea')
addFeatureValue(train, test, 'GarageArea', 'div', 'GarageQual')

addFeatureValue(train, test, 'GarageCars', 'multi', 'GarageArea')
addFeatureValue(train, test, 'GarageCars', 'div', 'GarageArea')
addFeatureValue(train, test, 'GarageArea', 'div', 'GarageCars')

addFeatureValue(train, test, 'OverallQual', 'multi', 'YearBuilt')
addFeatureValue(train, test, 'OverallQual', 'div', 'YearBuilt')
addFeatureValue(train, test, 'YearBuilt', 'div', 'OverallQual')

addFeatureValue(train, test, 'OverallQual', 'multi', 'YearRemodAdd')
addFeatureValue(train, test, 'OverallQual', 'div', 'YearRemodAdd')
addFeatureValue(train, test, 'YearRemodAdd', 'div', 'OverallQual')

addFeatureValue(train, test, 'BsmtUnfSF', 'multi', 'TotalBsmtSF')
addFeatureValue(train, test, 'BsmtUnfSF', 'div', 'TotalBsmtSF')
addFeatureValue(train, test, 'TotalBsmtSF', 'div', 'BsmtUnfSF')

addFeatureValue(train, test, 'FullBath', 'multi', 'TotalGrSF')
addFeatureValue(train, test, 'FullBath', 'div', 'TotalGrSF')
addFeatureValue(train, test, 'TotalGrSF', 'div', 'FullBath')

addFeatureValue(train, test, 'HalfBath', 'multi', 'TotalGrSF')
addFeatureValue(train, test, 'HalfBath', 'div', 'TotalGrSF')
addFeatureValue(train, test, 'TotalGrSF', 'div', 'HalfBath')

addFeatureValue(train, test, 'BedroomAbvGr', 'multi', 'TotalGrSF')
addFeatureValue(train, test, 'BedroomAbvGr', 'div', 'TotalGrSF')
addFeatureValue(train, test, 'TotalGrSF', 'div', 'BedroomAbvGr')

addFeatureValue(train, test, '1stFlrSF', 'multi', 'OverallQual')
addFeatureValue(train, test, '1stFlrSF', 'multi', 'YearBuilt')
addFeatureValue(train, test, '1stFlrSF', 'multi', 'YearRemodAdd')
addFeatureValue(train, test, '1stFlrSF', 'multi', 'FullBath')

addFeatureValue(train, test, 'GrLivArea', 'multi', 'OverallQual')
addFeatureValue(train, test, 'GrLivArea', 'multi', 'YearBuilt')
addFeatureValue(train, test, 'GrLivArea', 'multi', 'YearRemodAdd')
addFeatureValue(train, test, 'GrLivArea', 'multi', 'FullBath')

# 2020/03/28追加分
addFeatureValue(train, test, 'MasVnrType', 'multi', 'MasVnrArea')
addFeatureValue(train, test, 'MasVnrType', 'div', 'MasVnrArea')
addFeatureValue(train, test, 'MasVnrArea', 'div', 'MasVnrType')

addFeatureValue(train, test, 'LotShape', 'multi', 'LotFrontage')
addFeatureValue(train, test, 'LotShape', 'div', 'LotFrontage')
addFeatureValue(train, test, 'LotFrontage', 'div', 'LotShape')

addFeatureValue(train, test, 'LotShape', 'multi', 'LotArea')
addFeatureValue(train, test, 'LotShape', 'div', 'LotArea')
addFeatureValue(train, test, 'LotArea', 'div', 'LotShape')

addFeatureValue(train, test, 'LotShape', 'multi', 'YearBuilt')
addFeatureValue(train, test, 'LotShape', 'div', 'YearBuilt')
addFeatureValue(train, test, 'YearBuilt', 'div', 'LotShape')

addFeatureValue(train, test, 'LotArea', 'multi', 'YearBuilt')
addFeatureValue(train, test, 'LotArea', 'div', 'YearBuilt')
addFeatureValue(train, test, 'YearBuilt', 'div', 'LotArea')

addFeatureValue(train, test, 'LotFrontage', 'multi', 'YearBuilt')
addFeatureValue(train, test, 'LotFrontage', 'div', 'YearBuilt')
addFeatureValue(train, test, 'YearBuilt', 'div', 'LotFrontage')
```

### データベースから編集済の学習データ、テストデータを取得し、特徴量と目的変数に分ける

```python
# データベースから学習データ、テストデータを取得
train_featured = pd.read_sql("SELECT * FROM train_featured", con=engine)
test_featured = pd.read_sql("SELECT * FROM test_featured", con=engine)

# 学習データを特徴量と目的変数に分ける
train_x = train_featured.drop(['Id', 'SalePrice'], axis=1)
train_y = train_featured['SalePrice']

# テストデータは特徴量のみなので、そのままでよい
test_x = test_featured.drop(['Id'], axis=1)
test_id = test_featured['Id']
```

#### モデル作成とバリデーション
LightGBMを使用してモデルを作成します。

```python
printTime('モデルの作成開始')

va_pred_list = []
pred_list = []

# 学習データを学習データとバリデーションデータに分ける
kf = KFold(n_splits=4, shuffle=True, random_state=71)

for tr_idx, va_idx in kf.split(train_x):
    tr_x, va_x = train_x.iloc[tr_idx], train_x.iloc[va_idx]
    tr_y, va_y = train_y.iloc[tr_idx], train_y.iloc[va_idx]

    # 特徴量と目的変数をlightgbmのデータ構造に変換する
    lgb_train = lgb.Dataset(tr_x, tr_y)
    lgb_eval = lgb.Dataset(va_x, va_y)

    # ハイパーパラメータの設定
    params = {
        'boosting_type': 'gbdt',
        'objective': 'regression_l2',
        'metric': 'l2',
        'num_leaves': 40,
        'learning_rate': 0.05,
        'feature_fraction': 0.9,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'lambda_l2': 2,
    }

    # 作成する決定木の数を指定
    num_round = 100

    # 学習の実行
    # カテゴリ変数をパラメータで指定している
    # バリデーションデータもモデルに渡し、学習の進行とともにスコアがどう変わるかモニタリングする
    model = lgb.train(params,
                      lgb_train,
                      num_boost_round=num_round,
                      # categorical_feature=targets_category,
                      valid_names=['train', 'valid'],
                      valid_sets=[lgb_train, lgb_eval],
                      verbose_eval=0)

    # バリデーションデータでのスコアの確認
    va_pred = model.predict(va_x)
    va_pred_list.append(va_pred)

    # バリデーションデータでのスコア(真の値の対数と予測値の対数の二乗平均平方根誤差 (RMSE))を計算する
    rmse = np.sqrt(mean_squared_error(np.log(va_y), np.log(va_pred)))

    print(f'RMSE: {rmse:.4f}')

    # 結果の可視化
    sns.jointplot(
        va_y,
        va_pred,
        kind="reg",
        xlim=(0, 600000),
        ylim=(0, 600000)
    ).set_axis_labels("true", "pred")

    # 予測
    pred_list.append(model.predict(test_x))


va_pred_list = np.array(va_pred_list)
pred_list = np.array(pred_list)

# 提出用ファイルの作成
submission = pd.DataFrame({'Id': test_id, 'SalePrice': np.mean(pred_list, axis=0)})
submission.to_csv('/kaggle/output/submission_ensemble.csv', index=False)

printTime('モデルの作成終了')
```
